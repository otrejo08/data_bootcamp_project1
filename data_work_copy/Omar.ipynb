```python
import pandas as pd
import numpy as np

#import CSV
graduate_csv = "graduate_earnings_all.csv"

graduate_df = pd.read_csv(graduate_csv)
```

```python
graduate_df.head()
```

```python
graduate_df.shape
```

```python
#Create new DF with columns needed for analysis
reduced_graduate_df = graduate_df.loc[:, ["institution_name", "deglevl", "ciptitle","grad_cohort", "grad_cohort_label", "year_postgrad", "p25_earnings", "p50_earnings", "p75_earnings", "state"]]
reduced_graduate_df.head()
```

```python
#Remove null values
reduced_graduate_df = reduced_graduate_df.loc[(reduced_graduate_df["p25_earnings"] > 0)]
reduced_graduate_df.head(50)
```

```python
#Check for null values
reduced_graduate_df.isnull().any()
```

```python
#Rename columns
clean_df = reduced_graduate_df.rename(index=str, columns={"institution_name": "College/University", "deglevl": "Degree", "ciptitle": "Program", "grad_cohort": "Cohort", "year_postgrad": "Years Post Grad", "p25_earnings": "25th Percentile Earnings", "p50_earnings": "50th Percentile Earnings", "p75_earnings": "75th Percentile Earnings", "state": "State"})
```

```python
#Add Average Earnings column to simplify earnings analysis
clean_df['Average Earnings($)'] = clean_df[["25th Percentile Earnings", "50th Percentile Earnings", "75th Percentile Earnings"]].apply(np.mean, axis=1)
clean_df.loc[:,"Average Earnings($)"] = clean_df["Average Earnings($)"].astype(int).map("{:.2f}".format)
clean_df.head()


#Final clean DF
new_clean_df = clean_df.loc[:, ["State", "College/University", "Program", "Degree", "Cohort", "Years Post Grad", "Average Earnings($)"]]
new_clean_df.dtypes
```

```python
new_clean_df["Average Earnings($)"] = pd.to_numeric(new_clean_df["Average Earnings($)"])
new_clean_df.dtypes
```

```python
#We are only analyzing the data for Bachelor, Masters, and Doctorates degrees
#Filtered by Bachelors
bachelor_df = new_clean_df.loc[(new_clean_df["Degree"]=="Baccalaureate")]
bachelor_df.head(6)
```

```python
grouped_df = bachelor_df.groupby(['State', 'Program', 'College/University', 'Years Post Grad', 'Cohort', 'Average Earnings($)'])
grouped_df.count()

avg_earnings_total = pd.DataFrame(grouped_df["Average Earnings($)"].sum())
avg_earnings_total

bachelors_average_df = avg_earnings_total.sort_values("Average Earnings($)", ascending=False)
bachelors_average_df

bachelors_average_df.to_csv('bachelors_average.csv')
```

```python
#Highest Average Earnings Bachelor Degrees in TX
tx_bachelor_df = bachelor_df.loc[(new_clean_df["State"]=="TX")]
tx_bachelor_df.head()
```

```python
tx_grouped_df = tx_bachelor_df.groupby(['State', 'Program', 'College/University', 'Years Post Grad', 'Cohort', 'Average Earnings($)'])
tx_grouped_df.count()
```

```python
avg_earnings_total = pd.DataFrame(tx_grouped_df["Average Earnings($)"].sum())
avg_earnings_total

top_6_df = avg_earnings_total.sort_values("Average Earnings($)", ascending=False)
top_6_df

top_6_df.to_csv('tx_top_6.csv')
```
