```python
%matplotlib notebook
import matplotlib.pyplot as plt
import pandas as pd
import numpy as np
```

```python
tx_top_6_bachelor = "tx_top_6.csv"

tx_bachelor_df = pd.read_csv(tx_top_6_bachelor)
tx_bachelor_df.dtypes
```

```python
tx_bachelor_df.plot(kind="bar", x="Program", y="Average Earnings($)", grid=True,
              title="TX Top 6 Bachelor Degrees Average Earnings")
plt.show()
```

```python
plt.title("TX Top 6 Bachelors Average Earnings")
plt.xlabel("Programs")
plt.ylabel("Average Earnings ($)")
```
